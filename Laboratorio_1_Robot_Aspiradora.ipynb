```python
from IPython.display import display
import ipywidgets as widgets
import time

class Tablero:
    """
    La clase Tablero permite crear y mostrar un "tablero" interactivo basado en una tabla HTML.
    Se utiliza IPython.display junto con ipywidgets para renderizar el contenido en un entorno interactivo (como Jupyter Notebook).

    Atributos:
      out (ipywidgets.HTML): Widget HTML que muestra el tablero en pantalla.
      tamano_celda (tuple): Dimensiones de cada celda en píxeles, en el formato (alto, ancho). Valor predeterminado: (50, 50).
      n_celdas (tuple): Número de celdas en el tablero, en el formato (número de filas, número de columnas). Valor predeterminado: (5, 5).

    Métodos:
      __init__(tamano_celda, n_celdas):
          Inicializa el tablero creando el widget HTML y almacenando la configuración de tamaño y cantidad de celdas.
      dibujar(objetos):
          Genera y actualiza la visualización del tablero. Recibe una lista de objetos y, para cada celda, 
          comprueba si algún objeto tiene sus coordenadas (x, y) correspondientes a esa celda. 
          Si es así, coloca en dicha celda un div con el "emoticon" del objeto, rotado según su ángulo y con un tamaño de fuente determinado.
    """

    def __init__(self, tamano_celda=(50, 50), n_celdas=(5, 5)):
        """
        Constructor de la clase Tablero.
        
        Parámetros:
          tamano_celda (tuple, opcional): Dimensiones de cada celda (alto, ancho) en píxeles.
          n_celdas (tuple, opcional): Número total de celdas, definido como (filas, columnas).
          
        Acciones:
          - Crea un widget HTML para mostrar el tablero.
          - Muestra el widget en el entorno de IPython (por ejemplo, en Jupyter Notebook).
          - Almacena los parámetros de tamaño de celda y cantidad de celdas.
        """
        self.out = widgets.HTML()
        display(self.out)
        self.tamano_celda = tamano_celda
        self.n_celdas = n_celdas

    def dibujar(self, objetos):
        """
        Construye y muestra el tablero con los objetos proporcionados.

        Parámetros:
          objetos (list): Lista de objetos que se desean representar en el tablero. 
                          Cada objeto debe tener al menos los siguientes atributos:
                            - x: Coordenada horizontal (columna) en la cuadrícula.
                            - y: Coordenada vertical (fila) en la cuadrícula.
                            - angulo: Ángulo de rotación (en grados) para la representación visual del objeto.
                            - tamano_emoticon: Tamaño (en píxeles) para la fuente del "emoticon" o símbolo.
                            - emoticon: El símbolo o texto que se desea mostrar en la celda.
        
        Funcionamiento:
          1. Se define la estructura básica del tablero como una tabla HTML con borde.
          2. Se recorre cada fila (usando la cantidad de celdas en la primera dimensión de n_celdas).
          3. Para cada fila, se recorre cada columna (usando la segunda dimensión de n_celdas).
          4. Inicialmente, cada celda se crea vacía (variable 'contenido').
          5. Se recorre la lista de objetos y, si se encuentra un objeto cuya posición (o.x, o.y)
             coincide con las coordenadas actuales (columna j y fila i), se genera un <div> con:
                - Estilo de transformación para rotarlo según 'angulo'.
                - Tamaño de fuente definido por 'tamano_emoticon'.
                - Contenido del 'emoticon' a mostrar.
          6. Se construye la celda (<td>) asignándole el tamaño especificado (altura y ancho de tamano_celda) y se inserta el contenido generado.
          7. Se agrupan las celdas en filas (<tr>) y luego se insertan todas las filas en la tabla.
          8. Finalmente, se asigna el HTML generado al widget (self.out.value) para actualizar la visualización del tablero.
        """
        # Estructura base de la tabla HTML con borde.
        tablero = "<table border='1' >{}</table>"
        filas = ""

        # Iterar sobre las filas (la cantidad de filas se encuentra en self.n_celdas[0])
        for i in range(self.n_celdas[0]):
            s = ""  # 's' acumula el HTML de cada celda en la fila actual.
            # Iterar sobre las columnas (la cantidad de columnas se encuentra en self.n_celdas[1])
            for j in range(self.n_celdas[1]):
                contenido = ""  # Inicialmente, la celda está vacía.
                # Recorrer cada objeto para ver si debe aparecer en la celda (i, j)
                for o in objetos:
                    if o.x == j and o.y == i:
                        # Si el objeto se encuentra en la celda, se crea un <div> con su representación
                        contenido = (
                            "<div style='transform: rotate({angulo}deg);"
                            "font-size:{tamano_emoticon}px;'>{emoticon}</div>"
                        ).format(
                            angulo=o.angulo,
                            tamano_emoticon=o.tamano_emoticon,
                            emoticon=o.emoticon,
                        )
                # Se agrega la celda (<td>) con el contenido (y se define su tamaño usando tamano_celda)
                s += "<td style='height:{alto}px;width:{ancho}px'>{contenido}</td>".format(
                    alto=self.tamano_celda[0],
                    ancho=self.tamano_celda[1],
                    contenido=contenido,
                )
            # Se forma una fila de la tabla (<tr>) con todas las celdas acumuladas
            filas += "<tr>{}</tr>".format(s)
        # Se inserta el HTML de todas las filas en la estructura base de la tabla
        tablero = tablero.format(filas)
        # Se actualiza el widget HTML para mostrar el tablero generado
        self.out.value = tablero
```

```python
class AgenteReflejo(Agente):
    def __init__(self, x=0, y=0, angulo=0, emoticon="🧹", tamaño_emoticon=30, energia=5):
        """
        Inicializa los valores de la aspiradora
        """
        super().__init__(x, y, angulo, emoticon, tamaño_emoticon)
        self.energia = energia
        self.puntaje = 0

    def hayHoja(self, las_hojas):
        """
        Compara la posición actual de la aspiradora con la posición de todas las hojas en el tablero,
        devuelve True si hay una hoja en la misma posición de la aspiradora
        """
        aspiro = False
        for hoja in las_hojas:
            if hoja.x == self.x and hoja.y == self.y:
                las_hojas.remove(hoja)
                self.puntaje += 1
                aspiro = True
                break

        return aspiro
def percibir(self, las_hojas):
    """
    Percibe las hojas alrededor de la aspiradora y selecciona una al azar
    """
    percepcion = []

    for hoja in las_hojas:
        # Si la hoja está en el norte
        if hoja.x == self.x and hoja.y == self.y - 1:
            percepcion.append("N")

        # Si la hoja está en el sur
        if hoja.x == self.x and hoja.y == self.y + 1:
            percepcion.append("S")

        # Si la hoja está en el oeste
        if hoja.x == self.x - 1 and hoja.y == self.y:
            percepcion.append("O")

        # Si la hoja está en el este
        if hoja.x == self.x + 1 and hoja.y == self.y:
            percepcion.append("E")

if len(percepcion) > 0:
    direccion = int(np.random.random() * len(percepcion))
    return percepcion[direccion]
else:
    direcciones = int(np.random.random() * len(percepcion))
    return percepciones[direccion]

import numpy as np
import matplotlib.pyplot as plt

def crearAspiradora(n_celdas=(5, 5), energia=100):
    """
    Genera una posición aleatoria sobre el tablero en la cual se ubicará la aspiradora.

    Args:
        n_celdas: una tupla que establece el tamaño del tablero.
        energia: la cantidad de acciones que puede realizar la aspiradora.

    Return:
        Posición aleatoria en el tablero para ubicar la aspiradora.
    """
    # Posición aleatoria en el tablero para ubicar la aspiradora
    pos_x = int(np.random.random() * n_celdas[0])
    pos_y = int(np.random.random() * n_celdas[1])

    # Crear instancia de la aspiradora en la posición generada
    aspiradora = AgenteReflejo(pos_x, pos_y, energia=energia)

    return aspiradora

```
