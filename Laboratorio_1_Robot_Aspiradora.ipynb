```python
from IPython.display import display
import ipywidgets as widgets
import time

class Tablero:
    """
    La clase Tablero permite crear y mostrar un "tablero" interactivo basado en una tabla HTML.
    Se utiliza IPython.display junto con ipywidgets para renderizar el contenido en un entorno interactivo (como Jupyter Notebook).

    Atributos:
      out (ipywidgets.HTML): Widget HTML que muestra el tablero en pantalla.
      tamano_celda (tuple): Dimensiones de cada celda en p칤xeles, en el formato (alto, ancho). Valor predeterminado: (50, 50).
      n_celdas (tuple): N칰mero de celdas en el tablero, en el formato (n칰mero de filas, n칰mero de columnas). Valor predeterminado: (5, 5).

    M칠todos:
      __init__(tamano_celda, n_celdas):
          Inicializa el tablero creando el widget HTML y almacenando la configuraci칩n de tama침o y cantidad de celdas.
      dibujar(objetos):
          Genera y actualiza la visualizaci칩n del tablero. Recibe una lista de objetos y, para cada celda, 
          comprueba si alg칰n objeto tiene sus coordenadas (x, y) correspondientes a esa celda. 
          Si es as칤, coloca en dicha celda un div con el "emoticon" del objeto, rotado seg칰n su 치ngulo y con un tama침o de fuente determinado.
    """

    def __init__(self, tamano_celda=(50, 50), n_celdas=(5, 5)):
        """
        Constructor de la clase Tablero.
        
        Par치metros:
          tamano_celda (tuple, opcional): Dimensiones de cada celda (alto, ancho) en p칤xeles.
          n_celdas (tuple, opcional): N칰mero total de celdas, definido como (filas, columnas).
          
        Acciones:
          - Crea un widget HTML para mostrar el tablero.
          - Muestra el widget en el entorno de IPython (por ejemplo, en Jupyter Notebook).
          - Almacena los par치metros de tama침o de celda y cantidad de celdas.
        """
        self.out = widgets.HTML()
        display(self.out)
        self.tamano_celda = tamano_celda
        self.n_celdas = n_celdas

    def dibujar(self, objetos):
        """
        Construye y muestra el tablero con los objetos proporcionados.

        Par치metros:
          objetos (list): Lista de objetos que se desean representar en el tablero. 
                          Cada objeto debe tener al menos los siguientes atributos:
                            - x: Coordenada horizontal (columna) en la cuadr칤cula.
                            - y: Coordenada vertical (fila) en la cuadr칤cula.
                            - angulo: 츼ngulo de rotaci칩n (en grados) para la representaci칩n visual del objeto.
                            - tamano_emoticon: Tama침o (en p칤xeles) para la fuente del "emoticon" o s칤mbolo.
                            - emoticon: El s칤mbolo o texto que se desea mostrar en la celda.
        
        Funcionamiento:
          1. Se define la estructura b치sica del tablero como una tabla HTML con borde.
          2. Se recorre cada fila (usando la cantidad de celdas en la primera dimensi칩n de n_celdas).
          3. Para cada fila, se recorre cada columna (usando la segunda dimensi칩n de n_celdas).
          4. Inicialmente, cada celda se crea vac칤a (variable 'contenido').
          5. Se recorre la lista de objetos y, si se encuentra un objeto cuya posici칩n (o.x, o.y)
             coincide con las coordenadas actuales (columna j y fila i), se genera un <div> con:
                - Estilo de transformaci칩n para rotarlo seg칰n 'angulo'.
                - Tama침o de fuente definido por 'tamano_emoticon'.
                - Contenido del 'emoticon' a mostrar.
          6. Se construye la celda (<td>) asign치ndole el tama침o especificado (altura y ancho de tamano_celda) y se inserta el contenido generado.
          7. Se agrupan las celdas en filas (<tr>) y luego se insertan todas las filas en la tabla.
          8. Finalmente, se asigna el HTML generado al widget (self.out.value) para actualizar la visualizaci칩n del tablero.
        """
        # Estructura base de la tabla HTML con borde.
        tablero = "<table border='1' >{}</table>"
        filas = ""

        # Iterar sobre las filas (la cantidad de filas se encuentra en self.n_celdas[0])
        for i in range(self.n_celdas[0]):
            s = ""  # 's' acumula el HTML de cada celda en la fila actual.
            # Iterar sobre las columnas (la cantidad de columnas se encuentra en self.n_celdas[1])
            for j in range(self.n_celdas[1]):
                contenido = ""  # Inicialmente, la celda est치 vac칤a.
                # Recorrer cada objeto para ver si debe aparecer en la celda (i, j)
                for o in objetos:
                    if o.x == j and o.y == i:
                        # Si el objeto se encuentra en la celda, se crea un <div> con su representaci칩n
                        contenido = (
                            "<div style='transform: rotate({angulo}deg);"
                            "font-size:{tamano_emoticon}px;'>{emoticon}</div>"
                        ).format(
                            angulo=o.angulo,
                            tamano_emoticon=o.tamano_emoticon,
                            emoticon=o.emoticon,
                        )
                # Se agrega la celda (<td>) con el contenido (y se define su tama침o usando tamano_celda)
                s += "<td style='height:{alto}px;width:{ancho}px'>{contenido}</td>".format(
                    alto=self.tamano_celda[0],
                    ancho=self.tamano_celda[1],
                    contenido=contenido,
                )
            # Se forma una fila de la tabla (<tr>) con todas las celdas acumuladas
            filas += "<tr>{}</tr>".format(s)
        # Se inserta el HTML de todas las filas en la estructura base de la tabla
        tablero = tablero.format(filas)
        # Se actualiza el widget HTML para mostrar el tablero generado
        self.out.value = tablero
```

```python
class AgenteReflejo(Agente):
    def __init__(self, x=0, y=0, angulo=0, emoticon="游빛", tama침o_emoticon=30, energia=5):
        """
        Inicializa los valores de la aspiradora
        """
        super().__init__(x, y, angulo, emoticon, tama침o_emoticon)
        self.energia = energia
        self.puntaje = 0

    def hayHoja(self, las_hojas):
        """
        Compara la posici칩n actual de la aspiradora con la posici칩n de todas las hojas en el tablero,
        devuelve True si hay una hoja en la misma posici칩n de la aspiradora
        """
        aspiro = False
        for hoja in las_hojas:
            if hoja.x == self.x and hoja.y == self.y:
                las_hojas.remove(hoja)
                self.puntaje += 1
                aspiro = True
                break

        return aspiro
def percibir(self, las_hojas):
    """
    Percibe las hojas alrededor de la aspiradora y selecciona una al azar
    """
    percepcion = []

    for hoja in las_hojas:
        # Si la hoja est치 en el norte
        if hoja.x == self.x and hoja.y == self.y - 1:
            percepcion.append("N")

        # Si la hoja est치 en el sur
        if hoja.x == self.x and hoja.y == self.y + 1:
            percepcion.append("S")

        # Si la hoja est치 en el oeste
        if hoja.x == self.x - 1 and hoja.y == self.y:
            percepcion.append("O")

        # Si la hoja est치 en el este
        if hoja.x == self.x + 1 and hoja.y == self.y:
            percepcion.append("E")

if len(percepcion) > 0:
    direccion = int(np.random.random() * len(percepcion))
    return percepcion[direccion]
else:
    direcciones = int(np.random.random() * len(percepcion))
    return percepciones[direccion]

import numpy as np
import matplotlib.pyplot as plt

def crearAspiradora(n_celdas=(5, 5), energia=100):
    """
    Genera una posici칩n aleatoria sobre el tablero en la cual se ubicar치 la aspiradora.

    Args:
        n_celdas: una tupla que establece el tama침o del tablero.
        energia: la cantidad de acciones que puede realizar la aspiradora.

    Return:
        Posici칩n aleatoria en el tablero para ubicar la aspiradora.
    """
    # Posici칩n aleatoria en el tablero para ubicar la aspiradora
    pos_x = int(np.random.random() * n_celdas[0])
    pos_y = int(np.random.random() * n_celdas[1])

    # Crear instancia de la aspiradora en la posici칩n generada
    aspiradora = AgenteReflejo(pos_x, pos_y, energia=energia)

    return aspiradora

```
