```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/camilousa/datasets/refs/heads/master/german_credit_data(2)(1).csv")
df.head()
df.info()
df.describe()
```

```python
train, test = train_test_split(df, test_size=0.3)
```

```python
train.info()
test.info()
```

```python
train, test = train_test_split(df, test_size=0.3, random_state=42)
```

```python
x_train = train[["Age", "Credit_amount"]]
y_train = train["Risk"]
x_test = test[["Age", "Credit_amount"]]
y_test = test["Risk"]
```

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt

model = DecisionTreeClassifier()
model.fit(x_train, y_train)
plot_tree(model, filled=True)
plt.show()
```

```python
from sklearn.metrics import accuracy_score
model.fit(train[["Age", "Credit_amount"]], train["Risk"])
y_pred = model.predict(train[["Age", "Credit_amount"]])
accuracy = accuracy_score(train["Risk"], y_pred)
print(f"Accuracy: {accuracy}")
```

```python
#ahora hacer test
y_test_pred = model.predict(test[["Age", "Credit_amount"]])
accuracy = accuracy_score(test["Risk"], y_test_pred)
print(f"Accuracy: {accuracy}")

```
