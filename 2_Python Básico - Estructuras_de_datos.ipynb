## __Listas__
Las listas son una colección ordenada de elementos. Los elementos están identificados por un índice que va desde `0` hasta `n-1`, donde `n` es la cantidad de elementos de la lista. Las listas no requieren que sus elementos sean del mismo tipo.

```python
a = ['a', 'b', 'c', 'd']
b = [1, 2, 3, 4, 5, 6]
c = [1, 1.2, [3], True, 'hola', None]
```

Las listas  pueden tener a otras listas como elementos

```python
a = [1, 2 , [2, 3, 4, 5], True, 'hola']
a[4][2]
```

Podemos crear listas vacias de la siguiente forma:

```python
a = list()
b = []
print(a)
print(b)
```

También podemos usar el constructor `list()` para generar una nueva lista.

```python
a = list(("Manizales", "Bogotá", "Cali")) # note le doble parentesis
print(a)
```

Adicionalmente, también podemos definir anotaciones con el propósito de especificar un tipo de dato en particular, sin embargo, para el caso de las listas (y todas las colecciones) debemos utilizar el módulo `typing`

```python
from typing import List

a : List[int] = [1, 2, 3, 4]
b : List[bool] = [True, False, True]
c : List[int] = []

print(a)
print(b)
print(c)
```

Es posible particionar las listas (slicing) usando una notación del estilo `i:j`. Esto retornará una lista con los elementos desde la posición `i` a la `j-1`.

### __Ordenar y copiar listas__

Las listas cuentan con el método `sort()` que ordenará la lista de manera ascendente

```python
numeros = [100, 50, 65, 82, 23]
print(numeros)
numeros.sort()
print(numeros)
```

En el caso de cadenas de caracteres:

```python
a = list('Mercurio')
print(a)
a.sort()
print(a)
planetas = ['Mercurio', 'Tierra', 'Marte', 'Jupiter']
print(planetas)
planetas.sort()
print(planetas)
```

Si queremos ordenar de forma descendente podemos agregar el argumento `reverse = True  `

```python
numeros = [100, 50, 65, 82, 23]
numeros.reverse()
print(numeros)
numeros.sort(reverse=True)
print(numeros)

planetas = ['Mercurio', 'Tierra', 'Marte', 'Jupiter']
planetas.sort(reverse=True)
print(planetas)
```

El método `reverse()` nos permite invertir el orden de una lista

```python
planetas = ['Mercurio', 'Tierra', 'Marte', 'Jupiter']
planetas.reverse()
print(planetas)
```

Es importante considerar que los métodos `sort()` y `reverse()` no retornan una nueva lista, sino que modifican la lista actual.

```python
numeros = [100, 50, 65, 82, 23]

numeros_ordenados = numeros.sort()
numeros_invertidos = numeros.reverse()
print(numeros_ordenados)
print(numeros_invertidos)
print(numeros)
```

Para copiar una lista no podemos simplemente realizar una operación como `lista1 = lista2 ` debido a que `lista2` se convertirá en una referencia a `lista1` y cualquier cambio en `lista1` aparecerá automáticamente en `lista2`

```python
planetas = ['Mercurio', 'Tierra', 'Marte', 'Jupiter']

nuevos_planetas = planetas

nuevos_planetas[1:3] = ['a', 'e']

print(planetas)
print(nuevos_planetas)
```

Para evitar esto podemos utilizar el método `copy()` o el constructor `list()`

```python
planetas = ['Mercurio', 'Tierra', 'Marte', 'Jupiter']
nuevos_planetas = planetas.copy()
nuevos_planetas2 = list(planetas)
planetas[1:3] = ['a', 'e']

print(planetas)
print(nuevos_planetas)
print(nuevos_planetas2)
```

## __Tuplas__

Las tuplas funcionan igual que las listas, la principal diferencia es que las tuplas no son mutables (no pueden cambiar después de ser declarada)

```python
tupla1 = (1, 2, 3, 4)
tupla2 = 1, 2, 3, 4

print(type(tupla1))
print(tupla1)
print(tupla2)
print(tupla1[-1])
print(tupla1[:3])
print(tupla1[1:4])
```

También podemos utilizar anotaciones importando el módulo `typing`

```python
from typing import Tuple

tupla: Tuple[int] = (1, 2, 3, 4)
print(tupla)
```

Para crear tuplas con un único elemento debemo utilizar una coma `','`

```python
tupla1 = ("lunes",)
print(type(tupla1))

# No es una tupla
tupla2 = ("martes")
print(type(tupla2))
```

Una vez que una tupla es creada no es posible modificarla. Si necesitamos agregar un nuevo valor, o modificar alguno de los existentes, una alternativa es convertir la tupla a lista, hacer las modificaciones y generar una nueva tupla

```python
planetas = ('Mercurio', 'Tierra', 'Marte', 'Jupiter')
nuevos_planetas = list(planetas)

print(planetas)
nuevos_planetas.insert(1, 'Venus')
planetas = tuple(nuevos_planetas)

print(nuevos_planetas)
```

imagen.png

## __Conjuntos__

Los conjuntos son estructuras que permiten almacenar colecciones de datos con la propiedad de no poseer un orden, no tener elementos repetidos y no contar con un índice

```python
frutas = {'Manzana', 'peras', 'Limón', 'Manzana', 'Limón'}
C = {'Manzana', 2.0, True, False, True}

print(frutas)
print(C)
```

Un vez que un conjunto es creado no se pueden modificar sus elementos, sin embargo, es posible eliminar y agregar un elemento nuevo

```python
frutas = {'Manzana', 'peras', 'Limón', 'Manzana', 'Limón'}
print(frutas)
frutas.add('Naranja')
print(frutas)
frutas.remove('Manzana')
print(frutas)
```

### __Acceder a los elementos__

Por otro lado, para acceder a los elementos debemos iterar sobre el conjunto. No es posible hacer referencia a un elemento a través de un índice o una llave

```python
frutas = {'Manzana', 'peras', 'Limón', 'Manzana', 'Limón'}

for fruta in frutas:
  print(fruta)
```

### __Eliminar elementos__

Para eliminar un elemento podemos utilizar los métodos `remove()` o `discard()`. La diferencia es que el primero, en caso que el elemento no existe generará un error. El segundo no lo hará

```python
# Eliminar elemento
frutas = {'Manzana', 'peras', 'Limón'}
frutas.remove('peras')
print(frutas)

frutas = {'Manzana', 'peras', 'Limón'}
frutas.remove('Naranjas')
print(frutas)
```

```python
# Eliminar elemento
frutas = {'Manzana', 'peras', 'Limón'}
frutas.discard('Naranjas')
print(frutas)

frutas = {'Manzana', 'peras', 'Limón'}
frutas.discard('peras')
print(frutas)
```

### __Operaciones entre conjuntos__

Python cuenta con las operaciones básicas entre conjuntos

```python
# Operaciones entre conjuntos
frutas = {'Manzana', 'peras', 'Limón'}
A = {1, 2, 3}
B = {3, 4, 5}

# pertenencia
print("naranja" in frutas)
print("naranja" not in frutas)

# union
print(f'Unión: {A.union(B)}')

# intesección
print(f'Intersección: {A.intersection(B)}')

# diferencia
print(f'Diferencia: {A.difference(B)}')
```

## __Strings__

Los Strings o cadenas de caracteres son definidos como objetos iterables en Python

```python
a = "hola mundo"
b = '¿Hoy es lunes?'
print(type(a))
print(type(b))
print(a)
print(b)
```

Es posible acceder a los elementos de un String mediante el índice que indica la posición del caracter requerido

```python
a = "hola mundo"

print(a[0], a[-1])
print(a[1:-1])
```

Adicionalmente, podemos iterarlos a través de un ciclo:

```python
for char in 'hola Mundo':
  print(char, end=' ')
```

Los strings se pueden delimitar con comillas simples o dobles. Esto es útil si necesito incluir comillas simples en la cadena (encierro con dobles) o si quiero incluir comillas dobles (encierro con simples). También podemos utilizar cadenas de caracteres con múltiples líneas utilizando tres comillas simples (dobles)

```python
a = "Una comilla simple ' en mi string"
b = 'Una comilla doble " en mi string'

c = '''Lorem ipsum dolor sit amet,
consectetur adipiscing elit,
sed do eiusmod tempor incididunt
ut labore et dolore magna aliqua.'''

print(a)
print()
print(b)
print()
print(c)
```

También podríamos agregar las comillas utilizando el caracter de secuencia de escape `\'` y `\"` en lugar de `'` y `"`. Este caracter hace que la comilla reciba un tratamiento especial. Podemos encontrar otros usos para este caracter tales como `\n` para una nueva linea, `\t` para una tabulación, o `\\` si quiero que aparezca una sola barra invertida.


```python
print("hola\nmundo")
print("hola\tmundo")
print("hola\\mundo")
```

En el caso de que no nos interese el uso de la barra invertida para señalizar secuencias de escape, podemos agregar una `r` al principio de la cadena para indicarle al interprete de python que la ignore

```python
print(r'hola\nmundo')
```

A continuación se presentan ejemplos para algunos de los métodos disponibles en la manipulación de cadenas de caracteres:

```python
# Longitud de una cadena
a = "Hola Mundo!"
print(len(a))

# verificar la presencia de una sucadena
print('Hola' in 'Hola Mundo!')

# verificar la ausencia  de una sucadena
print('hoy' not in 'Hola Mundo!')
```

De mayúsculas a minúsculas y viceversa, así como la forma para eliminar espacios al principio y el final de la cadena.

```python
a = "Hola Mundo!"
# convertir a mayúsuclas
b = a.upper()
print(b)

# convertir a minúsculas
b = a.lower()
print(b)

# Eliminar espacios al principio y al final
a = "    Hola Mundo!   "
print(a, len(a))
b = a.strip()
print(b, len(b))
```

Remplazar una subcadena y dividir la cadena a partir de un caracter específico

```python
# remplazar una subcadena
a = "Hola Mundo!"
b = a.replace('!', '!!!!')
print(b)

# dividir a partir de los caracteres vacios
a = "Hola Mundo!"
b = a.split('o')
print(b)

# concatenar cadenas de caracteres
a = 'Hola'
b = 'Mundo'
c = '!'

nueva_cadena = a + ' ' + b + c
print(nueva_cadena)
```
