```python
#3.  ¿Cuál fue el mes más caluroso del año 2014?

import pandas as pd

# Cargar los datos
url = "https://raw.githubusercontent.com/jpospinalo/time-series/main/vic_elec.csv"
df = pd.read_csv(url)

# Convertir la columna Time a datetime
df['Time'] = pd.to_datetime(df['Time'], format='mixed')

# Filtrar datos para el año 2014
df_2014 = df[df['Time'].dt.year == 2014]

# Agrupar por mes y calcular la temperatura promedio
monthly_temps = df_2014.groupby(df_2014['Time'].dt.month)['Temperature'].mean()

# Encontrar el mes con la temperatura promedio más alta
hottest_month = monthly_temps.idxmax()
hottest_temp = monthly_temps.max()

# Convertir el número del mes a nombre
months = {
    1: 'Enero', 2: 'Febrero', 3: 'Marzo', 4: 'Abril',
    5: 'Mayo', 6: 'Junio', 7: 'Julio', 8: 'Agosto',
    9: 'Septiembre', 10: 'Octubre', 11: 'Noviembre', 12: 'Diciembre'
}
hottest_month_name = months[hottest_month]

# Imprimir resultado
print(f"El mes más caluroso de 2014 fue {hottest_month_name}")
print(f"Temperatura promedio: {hottest_temp:.2f}°C")
print("\nTemperaturas promedio por mes:")
for month, temp in monthly_temps.items():
    print(f"{months[month]}: {temp:.2f}°C")
```

```python
# 4.  ¿Cuál fue el mes más con mayor demanda de electricidad en el año 2014?

import pandas as pd

# Cargar los datos
url = "https://raw.githubusercontent.com/jpospinalo/time-series/main/vic_elec.csv"
df = pd.read_csv(url)

# Convertir la columna Time a datetime
df['Time'] = pd.to_datetime(df['Time'], format='mixed')

# Filtrar datos para el año 2014
df_2014 = df[df['Time'].dt.year == 2014]

# Agrupar por mes y calcular la demanda promedio
monthly_demand = df_2014.groupby(df_2014['Time'].dt.month)['Demand'].mean()

# Encontrar el mes con la demanda promedio más alta
highest_demand_month = monthly_demand.idxmax()
highest_demand = monthly_demand.max()

# Convertir el número del mes a nombre
months = {
    1: 'Enero', 2: 'Febrero', 3: 'Marzo', 4: 'Abril',
    5: 'Mayo', 6: 'Junio', 7: 'Julio', 8: 'Agosto',
    9: 'Septiembre', 10: 'Octubre', 11: 'Noviembre', 12: 'Diciembre'
}
highest_demand_month_name = months[highest_demand_month]

# Imprimir resultado
print(f"El mes con mayor demanda de electricidad en 2014 fue {highest_demand_month_name}")
print(f"Demanda promedio: {highest_demand:.2f} MW")
print("\nDemanda promedio por mes:")
for month, demand in monthly_demand.items():
    print(f"{months[month]}: {demand:.2f} MW")
```

```python
# 4. Determine en que tipo de día se consume más energía. Fines de semana, días laborales o festivos.

import pandas as pd

# Cargar los datos
url = "https://raw.githubusercontent.com/jpospinalo/time-series/main/vic_elec.csv"
df = pd.read_csv(url)

# Convertir la columna Time a datetime
df['Time'] = pd.to_datetime(df['Time'], format='mixed')

# Filtrar datos para el año 2014
df_2014 = df[df['Time'].dt.year == 2014]

# Agregar columnas para tipo de día
df_2014['Weekday'] = df_2014['Time'].dt.weekday  # 0 = Lunes, 6 = Domingo
df_2014['IsWeekend'] = df_2014['Weekday'].isin([5, 6])  # Sábado o Domingo
df_2014['IsWeekday'] = ~df_2014['IsWeekend'] & ~df_2014['Holiday']  # Laboral no festivo
df_2014['IsHoliday'] = df_2014['Holiday']

# Calcular demanda promedio por tipo de día
weekday_demand = df_2014[df_2014['IsWeekday']]['Demand'].mean()
weekend_demand = df_2014[df_2014['IsWeekend'] & ~df_2014['Holiday']]['Demand'].mean()
holiday_demand = df_2014[df_2014['IsHoliday']]['Demand'].mean()

# Imprimir resultados
print(f"Demanda promedio por tipo de día en 2014:")
print(f"Días laborales: {weekday_demand:.2f} MW")
print(f"Fines de semana: {weekend_demand:.2f} MW")
print(f"Festivos: {holiday_demand:.2f} MW")

# Determinar el tipo de día con mayor consumo
max_demand = max(weekday_demand, weekend_demand, holiday_demand)
if max_demand == weekday_demand:
    print("\nLos días laborales tienen el mayor consumo de energía.")
elif max_demand == weekend_demand:
    print("\nLos fines de semana tienen el mayor consumo de energía.")
else:
    print("\nLos días festivos tienen el mayor consumo de energía.")
```

```python
# 5. ¿Existe alguna correlación significativa entre la temperatura y la demanda de energía? Utilice el coeficiente de correlación de Pearson y realice un diagrama de dispersión.
import pandas as pd

# Cargar los datos
url = "https://raw.githubusercontent.com/jpospinalo/time-series/main/vic_elec.csv"
df = pd.read_csv(url)

# Convertir la columna Time a datetime
df['Time'] = pd.to_datetime(df['Time'], format='mixed')

# Filtrar datos para el año 2014
df_2014 = df[df['Time'].dt.year == 2014]

# Calcular demanda promedio para días festivos y no festivos
holiday_demand = df_2014[df_2014['Holiday'] == True]['Demand'].mean()
non_holiday_demand = df_2014[df_2014['Holiday'] == False]['Demand'].mean()

# Calcular la diferencia (no festivos - festivos)
demand_difference = non_holiday_demand - holiday_demand

# Imprimir resultados
print(f"Demanda promedio en días festivos: {holiday_demand:.2f} MW")
print(f"Demanda promedio en días no festivos: {non_holiday_demand:.2f} MW")
print(f"Diferencia promedio (no festivos - festivos): {demand_difference:.2f} MW")

# Interpretar el resultado
if demand_difference > 0:
    print(f"En promedio, los días no festivos consumen {demand_difference:.2f} MW más que los festivos.")
elif demand_difference < 0:
    print(f"En promedio, los días festivos consumen {abs(demand_difference):.2f} MW más que los no festivos.")
else:
    print("No hay diferencia promedio en la demanda entre días festivos y no festivos.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
import numpy as np

# ¿Existe alguna correlación significativa entre la temperatura y la demanda de energía? 
# Utilice el coeficiente de correlación de Pearson y realice un diagrama de dispersión.

# Cargar los datos
url = "https://raw.githubusercontent.com/jpospinalo/time-series/main/vic_elec.csv"
df = pd.read_csv(url)

# Convertir la columna Time a datetime
df['Time'] = pd.to_datetime(df['Time'], format='mixed')

# Filtrar datos para el año 2014
df_2014 = df[df['Time'].dt.year == 2014]

# Calcular el coeficiente de correlación de Pearson
correlation, p_value = pearsonr(df_2014['Temperature'], df_2014['Demand'])

# Imprimir resultados
print(f"Coeficiente de correlación de Pearson: {correlation:.3f}")
print(f"Valor p: {p_value:.3e}")
if p_value < 0.05:
    print("La correlación es estadísticamente significativa (p < 0.05)")
else:
    print("La correlación no es estadísticamente significativa (p >= 0.05)")

# Crear diagrama de dispersión
plt.figure(figsize=(10, 6))
plt.scatter(df_2014['Temperature'], df_2014['Demand'], alpha=0.1, s=10)
plt.title('Relación entre Temperatura y Demanda de Energía - 2014')
plt.xlabel('Temperatura (°C)')
plt.ylabel('Demanda (MW)')
plt.grid(True)

# Agregar línea de tendencia
z = np.polyfit(df_2014['Temperature'], df_2014['Demand'], 1)
p = np.poly1d(z)
plt.plot(df_2014['Temperature'], p(df_2014['Temperature']), "r--", alpha=0.5)

plt.show()
```

```python
#6. Realice nuevamente un gráfico de dispersión como en el punto anterior, pero ahora visualizando los valores de la columna `Day_type` cada uno en un color diferente. ¿Qué conclusiones podemos obtener a partir del gráfico?
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
import numpy as np

# Cargar los datos
url = "https://raw.githubusercontent.com/jpospinalo/time-series/main/vic_elec.csv"
df = pd.read_csv(url)

# Convertir la columna Time a datetime
df['Time'] = pd.to_datetime(df['Time'], format='mixed')

# Filtrar datos para el año 2014
df_2014 = df[df['Time'].dt.year == 2014]

# Crear clasificación de tipos de día
df_2014['Weekday'] = df_2014['Time'].dt.weekday  # 0 = Lunes, 6 = Domingo
df_2014['Day_type'] = 'Weekday'  # Valor por defecto
df_2014.loc[df_2014['Weekday'].isin([5, 6]) & ~df_2014['Holiday'], 'Day_type'] = 'Weekend'
df_2014.loc[df_2014['Holiday'], 'Day_type'] = 'Holiday'

# Crear diagrama de dispersión con colores por tipo de día
plt.figure(figsize=(12, 8))
for day_type, color in zip(['Weekday', 'Weekend', 'Holiday'], ['blue', 'green', 'red']):
    subset = df_2014[df_2014['Day_type'] == day_type]
    plt.scatter(subset['Temperature'], subset['Demand'], 
                alpha=0.1, s=10, c=color, label=day_type)

plt.title('Relación entre Temperatura y Demanda por Tipo de Día - 2014')
plt.xlabel('Temperatura (°C)')
plt.ylabel('Demanda (MW)')
plt.grid(True)
plt.legend(title='Tipo de Día')

# Agregar línea de tendencia general (opcional)
z = np.polyfit(df_2014['Temperature'], df_2014['Demand'], 1)
p = np.poly1d(z)
plt.plot(df_2014['Temperature'], p(df_2014['Temperature']), "k--", alpha=0.5)

plt.show()

# Calcular correlación por tipo de día (opcional)
for day_type in ['Weekday', 'Weekend', 'Holiday']:
    subset = df_2014[df_2014['Day_type'] == day_type]
    corr, p = pearsonr(subset['Temperature'], subset['Demand'])
    print(f"Correlación para {day_type}: {corr:.3f} (p = {p:.3e})")
```
